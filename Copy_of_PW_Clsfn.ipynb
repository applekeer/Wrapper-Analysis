```python
import cv2
import easyocr
import pandas as pd
from fuzzywuzzy import fuzz
from google.colab import files

# Upload image
uploaded = files.upload()

# Get the uploaded image file name
image_path = list(uploaded.keys())[0]

# Read image
img = cv2.imread(image_path)

# Convert BGR image to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Initialize EasyOCR reader
reader = easyocr.Reader(['en'], gpu=False)

# Read text from image
results = reader.readtext(img_rgb)

# Extract OCR-detected texts from valid results
detected_texts = [result[1] for result in results if result[1]]

# Sample data
data = {
    'Ingredient': ['Edible Vegetable Fat', 'Milk Solids', 'Cocoa Solids', 'Cashew Nuts', 'Emulsifier E322', 'Milk',
                   'Sugar', 'Cane Sugar', 'Honey Powder', 'Salt', 'Refined Wheat Flour', 'Ragi Flour', 'Refined Oil',
                   'Ghee', 'Iodized Salt', 'Millet', 'Artificial Fragrances', 'Raising Agent', 'Flavoring Substances',
                   'Citric Acid', 'Potato', 'Refined Palmolein Oil', 'Bengal Gram Flour (Besan)', 'Potato Flakes', 'Potato Starch',
                   'Tapioca Starch', 'Iodized Salt', 'Red Chilli Powder', 'Black Salt', 'Acidity Regulator',
                   'Black Pepper', 'Clove', 'Cardamom', 'Dried Ginger Powder', 'Dried Garlic Powder', 'Refined Sugar',
                   'Cumin Powder', 'Bay Leaves', 'Nutmeg', 'Cinnamon', 'Mint Leaves'],
    'Safety_Level': ['Safe', 'Safe (in moderate amounts)', 'Safe', 'Safe', 'Safe', 'Safe', 'Unsafe', 'Unsafe',
                     'Safe', 'Safe (in moderate amounts)', 'Unsafe', 'Safe', 'Safe (in moderate amounts)',
                     'Safe (in moderate amounts)', 'Safe (in moderate amounts)', 'Safe', 'Unsafe', 'Unsafe', 'Unsafe',
                     'Safe', 'Safe', 'Safe (in moderate amounts)', 'Safe', 'Safe', 'Safe', 'Safe', 'Safe (in moderate amounts)',
                     'Safe', 'Safe', 'Safe', 'Safe', 'Safe', 'Safe', 'Safe', 'Safe', 'Unsafe', 'Safe', 'Safe',
                     'Safe', 'Safe', 'Safe']
}

# Ensure both lists have the same length
assert len(data['Ingredient']) == len(data['Safety_Level']), "Lengths of 'Ingredient' and 'Safety_Level' lists must be equal"

# Create a DataFrame from the sample data
safety_df = pd.DataFrame(data)

# Function to classify OCR extracted text
def classify_ingredient(ingredient):
    for index, row in safety_df.iterrows():
        if fuzz.partial_ratio(ingredient.lower(), row['Ingredient'].lower()) >= 80:
            return row['Safety_Level']
    return "Not Found"


# Create a DataFrame from the sample data
safety_df = pd.DataFrame(data)

# Function to classify OCR extracted text
def classify_ingredient(ingredient):
    for index, row in safety_df.iterrows():
        if fuzz.partial_ratio(ingredient.lower(), row['Ingredient'].lower()) >= 80:
            return row['Safety_Level']

# Classify OCR extracted texts
classified_ingredients = {text: classify_ingredient(text) for text in detected_texts}

# Print classified ingredients
print("OCR Extracted Text\t\t\t\tSafety Level")
for text, safety_level in classified_ingredients.items():
    print(f"{text}: {safety_level}")
```
